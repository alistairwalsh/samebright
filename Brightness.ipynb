## How do you want to calculate this?

 I can 

1. get an average R value, G values and B value for the entire image. Then Average these three values.

2. Average the RGB values for each pixel, then average across all the pixels.

3. Just add all the values and average once. Is that the same as 1 or not? 

Each method has an advantages I suppose. For an overall brightness value, I think I'd average over a pixel then average across the image.

```python
(R+G+B)/3 ## Get an average
```

```python
brightness = sum([R,G,B])/3 ##0 is dark (black) and 255 is bright (white)
```

# Allowing for the different perception of the brightness of R, G and B.

This is the more complex calculation due to the difference in perceived brightness for different colours. It's probably not necessary. The values aren't wildly different from the basic averaging above 

```python
#Standard
LuminanceA = (0.2126*R) + (0.7152*G) + (0.0722*B)
#Percieved A
LuminanceB = (0.299*R + 0.587*G + 0.114*B)
#Perceived B, slower to calculate
LuminanceC = sqrt( 0.241*R**2 + 0.691*G**2 + 0.068*B**2 )
```
